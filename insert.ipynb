```python
import json
import os
import glob
```

```python
file_path = '/opt/ml/jsons'
file_list = glob.glob(file_path+'/*')
original_anno = '/opt/ml/input/dataset/annotation.json'
cnt = 0

for annotation_path in file_list:
    img_path = annotation_path.split('/')[-1][:-5]+'.jpg'

    if cnt >= 1:
        original_anno = '/opt/ml/new.json'

    with open(original_anno,'r') as f:
        origin = json.load(f)
    

    new_annotations = []
    with open (annotation_path,'r') as f:
        annotation = json.load(f)

    for index in annotation['shapes']:
        index['language'] = []
        language=''
        if index['shape_type'] == 'rectangle':
            a = index['points'][0][0]
            b = index['points'][0][1]
            c = index['points'][1][0]
            d = index['points'][1][1]
            index['points'] = [[a,b],[c,b],[c,d],[a,d]]
        for i in index['label']:
            if i.encode().isalpha():
                language = 'en'
                if language not in index['language']:
                    index['language'].append(language)
            elif i.isalpha():
                language = 'ko'
                if language not in index['language']:
                    index['language'].append(language)
            else:
                language = 'others'
                if language not in index['language']:
                    index['language'].append(language)  
        
        if index['label'] == 'null':
            index['illegibility'] = True
        else:
            index['illegibility'] = False

        new = {
            'transcription' : index['label'],
            'illegibility': index['illegibility'],
            'orientation': 'null',
            'language': index['language'],
            'word_tags': 'null',
            'points': index['points']
        }
        new_annotations.append(new)

    #print(annotation['shapes'][0])
    print(len(new_annotations))

    #print(origin['images']['1632655330045-20.jpg']['words'])

    words = origin['images'][f'{img_path}']['words']

    #print(words)

    idx = len(words)
    print(idx)

    for i in new_annotations:
        words[f'{idx}'] = i
        idx += 1
    print(len(words))
    with open('new.json','w') as f:
        json.dump(origin,f)
    cnt += 1
```
